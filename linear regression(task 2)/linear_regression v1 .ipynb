```python
#df.cut=df['cut'].astype('category')
#df.cut=df['cut'].cat.codes
one_hot = pd.get_dummies(df['cut'])
df = pd.concat([df, one_hot], axis=1)
```

```python
df.cut.value_counts()
```

```python
cpp=df.price/df.carat
df['cpp']=cpp
```

```python
min=df.cpp.quantile(.015)
max=df.cpp.quantile(.985)
#max_threshold
```

```python
df.corr()
```

```python
X = add_constant(df)  
ds=pd.Series([variance_inflation_factor(X.values, i)   
               for i in range(X.shape[1])],   
              index=X.columns)  
ds
```

```python
x=df[(df.cpp>min) & (max<df.cpp)].drop(columns='price').drop(columns='cpp').drop(columns='depth').drop(columns='cut').drop(columns='color').drop(columns='clarity')
y=df[(df.cpp>min) & (max<df.cpp)].price
#x=df.drop(columns='price').drop(columns='depth').drop(columns='cut').drop(columns='color').drop(columns='clarity')
#std=df.cpp.std()
#mean=df.cpp.mean()
#y=df.price
#x=df[(df.cpp<2.175*std+mean) & (df.cpp>mean- 1.8*std)].drop(columns='price').drop(columns='cpp')
#y=df[(df.cpp<2.175*std+mean) & (df.cpp>mean- 1.8*std)].price
```

```python
import sklearn 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
print(lr.score(x_test, y_test))
```
